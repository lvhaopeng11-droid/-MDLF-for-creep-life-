```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

# --- Set Global Font Style ---
plt.rcParams['font.family'] = 'Arial'

# --- Simulate Datasets ---
np.random.seed(42)
num_total_samples = 134
num_train_samples = 100
num_test_samples = num_total_samples - num_train_samples

# Define reasonable ranges for the features
min_Ad, max_Ad = 2.58, 5.23

# --- Construct Training Set Data ---
train_data = {
    'Ad': np.random.uniform(min_Ad, max_Ad, num_train_samples)
}
train_df = pd.DataFrame(train_data)

# --- Construct Test Set Data ---
test_data = {
    'Ad': np.random.uniform(min_Ad + 0.03, max_Ad - 0.03, num_test_samples)
}
test_df = pd.DataFrame(test_data)

# --- Plotting Section ---
fig, ax = plt.subplots(figsize=(9, 6), dpi=600)

# Plot histograms with darker KDE lines
sns.histplot(train_df['Ad'], kde=True, color='#FFE0C1', label='Training Set',
             ax=ax, stat='density', alpha=0.9, bins=22, zorder=2,
             line_kws={'linewidth': 2, 'color': '#EAB16E'})  # 深橙色KDE线，加粗

sns.histplot(test_df['Ad'], kde=True, color='#C9DEF9', label='Test Set',
             ax=ax, stat='density', alpha=0.9, bins=22, zorder=2,
             line_kws={'linewidth': 2, 'color': '#80BEB3'})  # 深蓝色KDE线，加粗

# ========== 新增部分：手动修改KDE线的颜色 ==========
# 获取所有线条对象
lines = ax.get_lines()

# 通常最后两条线是KDE线
# 第一条KDE线（Training Set）改为 #EAB16E
if len(lines) >= 2:
    lines[-2].set_color('#EAB16E')
    lines[-2].set_linewidth(2)

# 第二条KDE线（Test Set）改为 #80BEB3
if len(lines) >= 1:
    lines[-1].set_color('#80BEB3')
    lines[-1].set_linewidth(2)
# ========== 新增部分结束 ==========

# Set labels
ax.set_xlabel('Average $δ$ Phase Volume Fraction', fontsize=28, fontweight='bold', labelpad=20)
ax.set_ylabel('Density', fontsize=28, fontweight='bold', labelpad=20)

# --- 设置纵坐标范围和刻度 ---
ax.set_ylim(0.2, 1.4)
ax.set_yticks(np.arange(0.2, 1.41, 0.2))

# Style spines
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(2.0)

# Style tick labels
ax.tick_params(axis='x', which='major', labelsize=24)
ax.tick_params(axis='y', which='major', labelsize=24)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

# --- Set Global Font Style ---
plt.rcParams['font.family'] = 'Arial'

# --- Simulate Datasets ---
np.random.seed(42)
num_total_samples = 134
num_train_samples = 100
num_test_samples = num_total_samples - num_train_samples

# Define reasonable ranges for the features
min_D, max_D = 4.32, 39.71

# --- Construct Training Set Data ---
train_data = {
    'D': np.random.uniform(min_D, max_D, num_train_samples)
}
train_df = pd.DataFrame(train_data)

# --- Construct Test Set Data ---
test_data = {
    'D': np.random.normal((min_D + max_D) / 2, (max_D - min_D) / 8, num_test_samples)
}
test_df = pd.DataFrame(test_data)

# Clip the grain size data
test_df['D'] = np.clip(test_df['D'], min_D, max_D)

# --- Plotting Section ---
fig, ax = plt.subplots(figsize=(9, 6), dpi=600)

# Plot histograms
sns.histplot(train_df['D'], kde=True, color='#FFE0C1', label='Training Set', ax=ax, stat='density', alpha=0.7, bins=22, zorder=2)
sns.histplot(test_df['D'], kde=True, color='#C9DEF9', label='Test Set', ax=ax, stat='density', alpha=0.7, bins=22, zorder=2)
# ========== 新增部分：手动修改KDE线的颜色 ==========
# 获取所有线条对象
lines = ax.get_lines()

# 通常最后两条线是KDE线
# 第一条KDE线（Training Set）改为 #EAB16E
if len(lines) >= 2:
    lines[-2].set_color('#EAB16E')
    lines[-2].set_linewidth(2)

# 第二条KDE线（Test Set）改为 #80BEB3
if len(lines) >= 1:
    lines[-1].set_color('#80BEB3')
    lines[-1].set_linewidth(2)
# ========== 新增部分结束 ==========
# Set labels
ax.set_xlabel('Average Grain Size', fontsize=28, fontweight='bold',labelpad=20)
ax.set_ylabel('Density', fontsize=28, fontweight='bold',labelpad=20)
# --- 设置纵坐标范围和刻度 ---
ax.set_ylim(0,0.2)
ax.set_yticks(np.arange(0, 0.21, 0.05))
# Style spines
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(2.0)

# Style tick labels
ax.tick_params(axis='x', which='major', labelsize=24)
ax.tick_params(axis='y', which='major', labelsize=24)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# --- Font settings ---
plt.rcParams['font.family'] = ['Times New Roman']
# Enable display of negative signs
plt.rcParams['axes.unicode_minus'] = False

# --- Data definition ---
labels = ['D', 'Aδ', 'Lδ']
values = [6.60, 4.30, 1.96]  # [D, Aδ, Lδ]

# --- Colors and styling ---
bar_colors = ['#FFE0C1', '#C9DEF9', '#B0B7DB']  # distinct colors for each feature
edge_color = 'black'
edge_width = 1.5

# --- Create figure and axes ---
fig, ax1 = plt.subplots(figsize=(8, 6), dpi=600)

# 设置背景为白色（删除渐变背景）
fig.patch.set_facecolor('white')
ax1.set_facecolor('white')

# Left Y-axis for sizes (D and Lδ)
ax1.set_xlabel('Feature Name', fontsize=26, fontweight='bold',labelpad=20)
ax1.set_ylabel('Size (μm)', fontsize=26, fontweight='bold',labelpad=20)
ax1.set_ylim(0, 8)  # 设置范围为0到8

# Plot D and Lδ bars on left axis (with edge)
bars_D = ax1.bar(labels[0], values[0],
                 color=bar_colors[0],
                 edgecolor=edge_color, linewidth=edge_width,
                 width=0.5, label='D (Grain Size)',
                 zorder=3, alpha=0.9)
bars_Ld = ax1.bar(labels[2], values[2],
                  color=bar_colors[2],
                  edgecolor=edge_color, linewidth=edge_width,
                  width=0.5, label='Lδ (Lath Width)',
                  zorder=3, alpha=0.9)

# Right Y-axis for volume fraction (Aδ)
ax2 = ax1.twinx()
ax2.set_ylabel('Volume Fraction (%)', fontsize=26, fontweight='bold',labelpad=20)
ax2.set_ylim(0, 5)  # 设置范围为0到5

bars_Ad = ax2.bar(labels[1], values[1],
                  color=bar_colors[1],
                  edgecolor=edge_color, linewidth=edge_width,
                  width=0.5, label='Aδ (Area Fraction)',
                  zorder=3, alpha=0.9)

# --- Customize tick labels ---
for ax in (ax1, ax2):
    ax.tick_params(axis='y', labelsize=24)
    for lbl in ax.get_yticklabels():
        lbl.set_fontweight('bold')

ax1.tick_params(axis='x', labelsize=24)
for lbl in ax1.get_xticklabels():
    lbl.set_fontweight('bold')

# --- Add value labels atop bars ---
def add_value_labels(bars, ax):
    for bar in bars:
        h = bar.get_height()
        ax.annotate(f'{h:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, h),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom',
                    fontsize=26, fontweight='bold',
                    zorder=5)

add_value_labels(bars_D, ax1)
add_value_labels(bars_Ld, ax1)
add_value_labels(bars_Ad, ax2)

# --- Title and legend (with bold text) ---
handles = [bars_D, bars_Ad, bars_Ld]
legend_labels = ['D (Grain Size)', 'Aδ (Area Fraction)', 'Lδ (Lath Width)']
legend = ax1.legend(handles, legend_labels, loc='upper center', 
                    fontsize=24, 
                    prop={'family': 'Times New Roman', 'weight': 'bold'})

# 确保图例框架显示
legend.get_frame().set_linewidth(1.5)
legend.get_frame().set_edgecolor('black')
legend.get_frame().set_facecolor('white')
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(2.0)

# --- Final layout ---
fig.tight_layout()
plt.show()
```
